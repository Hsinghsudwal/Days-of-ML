# Understanding Your Data

## Asking the Simple Questions

We are using the titanic dataset for this example

```python
import pandas as pd
df = pd.read_csv('train.csv')
```

### 1. How Big is the Data?
The best way to do this is using the .shape

```python
df.shape
```

### 2. How does the Data Look?
The best way to do this is to read a few rows. Using .head()

```python
df.head()
```

```python
# sometimes you might need to know any random data not just the first 5 rows
df.sample(5) # will return 5 smaples from the dataframe
```

### 3. What is the DataType of the Columns?
The .info() method is used for this.

```python
df.info()
# sometimes it might be useful to convert a few datatypes beforehand to reduce computation lag in the future
# Eg: chnage age(float) to age(int)
```

### 4. Are there any missing values??
This is already visible in .info() function. If the Non-Null Count is not the same, then it indicates there are Null values.

```python
# this can be better evaluated when we use
df.isnull().sum()
# retuens a pandas series of column names and no of null values
```

### 5. How does the data look Mathematically?
Just use describe. It will return the mean, median, std, 25%, 50% atc mathematical analysis on Numerical Columns

```python
df.describe()
```

### 6. Are there any Duplicate Values?
Similar to missing values we have an inbuilt finction.

```python
df.duplicated().sum()
# returns a series of column names and number of duplicated values
# in this case all the rows are unique
```

### 7. How is the Correlation between Columns?
We sometimes have more columns than we need. In some analysis it might be a better idea to totally ignore the un-correlated columns. This ensures that they dont cause a bias in the model.

This also only works on numberical values.

```python
df.corr(numeric_only=True)
# will return a grid of all the numerical columns
```

```python
# there are cases when you need to find the correlation of one column with the rest of the columns
df.corr(numeric_only=True)['Survived']
# note the high correlation between fare and Survivability
# also note that it is negativaely correlated with Pclass, which is also a real world scenario
```
