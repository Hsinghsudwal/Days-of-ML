### SQL

For working with an SQL database with your local host. You need to get this data in your local host.

- in sql server, if you have it setup from VSCode, connect to the LocalHost
- open the SQL file you downloaded from Kaggle
- select everything and press "Ctrl+E, Ctrl+E"
- this will upload everything to you LocalHost

Now for connecting Python and SQL(LocalHost), we need some sort of bridge. We use the MySQL Connector (library) to do this. 

Install it to your Conda Environment, and then import it.

In my case, it was a required thing to change the authentication plugin.

```python
import mysql.connector
```

```python
# use this connector to get a connection object!
# make sure to change the "Authentication Plugin" to 'mysql_native_password'
conn = mysql.connector.connect(host='localhost', password='1234', user='throne', database='world', auth_plugin='mysql_native_password')
```

```python
# once the connection is established
# pass the SQL query as a string with the 'conn' object
pd.read_sql_query("select * from city", conn)
```

```python
# you can ply with your queries in this way
pd.read_sql_query("select * from city where CountryCode like 'IND'", conn)
```

```python
pd.read_sql_query("select * from country where LifeExpectancy > 60", conn)
# parameters like index= , parsedates= , chunksize=, etc are available here too
```
