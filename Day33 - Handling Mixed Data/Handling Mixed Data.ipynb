## Handling Mixed Data

```python
import numpy as np
import pandas as pd
```

```python
# this is modified Titanic Dataset
df = pd.read_csv('titanic.csv')
df.head()
```

```python
# let's see what unique values are present
df['number'].unique()
# this is columns made from sibling-spouse and parent-child and making 0 as 'A' (alone)
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers Travelling with')
```

```python
# extract numerical part
df['number_numerical'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
# converts to numbers, if there are errors force them to change and downcast the result to integer
df['number_categorical'] = np.where(df['number_numerical'].isnull(), df['number'], np.nan)
# np.where yeilds x when true and y when false

df.head()
```

```python
# now for the columns 'Cabin' and 'Ticket'
```

```python
df['Cabin'].unique()
# damn it has a lot of unique values
```

```python
# its the same with ticket
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # this is regular expression which filters numbers from the cabin column
df['cabin_cat'] = df['Cabin'].str[0]               # captures the first letter from the column

df.head()
```

```python
# now look at the categories we have!!
df['cabin_cat'].value_counts()
```

```python
# for the ticket number column
# numeric part
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1]) # split when there is white space " " in the value
df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors='coerce', downcast='integer')
# further forcibly change all the errors and downcast into integers

# categorical part
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0]) # split at white space and take the first part
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])
# if TRUE set NULL, else set df['Ticket']

df.head()
```

```python
# check out the unique categories from ticket_cat
df['ticket_cat'].unique()
# lot less than before
```
